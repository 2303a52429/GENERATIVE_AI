 (1 ponto) Write Python code without using any libraries to find the value of x at which the
 function f(x) shown in equation (1) has minimum value. Consider Gradient Descent Algorithm.
 f(x) = 5x4 +3x2 +1

```python
def derivative(x):
    return 20 * x**3 + 6 * x
def gradient_descent(initial_x, learning_rate, tolerance, max_iterations):
    x = initial_x
    for iteration in range(max_iterations):
        grad = derivative(x)
        new_x = x - learning_rate * grad

        print(f"Iteration {iteration + 1}: x = {new_x}, Gradient = {grad}")

        if abs(new_x - x) < tolerance:
            break

        x = new_x

    return x
initial_x = 1.0
learning_rate = 0.01
tolerance = 1e-6
max_iterations = 10000
optimal_x = gradient_descent(initial_x, learning_rate, tolerance, max_iterations)
print("The minimum value of x is approximately:", optimal_x,)
```

(1 ponto) Write Python code without using any libraries to find the value of x and y at which the
 function g(x,y) shown in equation (2) has minimum value. Consider Gradient Descent Algorithm.
 f(x) = 3x2 +5eâˆ’y +1

```python
def partial_derivatives(x, y):
    df_dx = 6 * x
    df_dy = -5 * (2.71828 ** -y)
    return df_dx, df_dy

def gradient_descent(initial_x, initial_y, learning_rate, tolerance, max_iterations):
    x, y = initial_x, initial_y
    for iteration in range(max_iterations):
        grad_x, grad_y = partial_derivatives(x, y)
        new_x = x - learning_rate * grad_x
        new_y = y - learning_rate * grad_y

        print(f"Iteration {iteration + 1}: x = {new_x}, y = {new_y}, Gradient_x = {grad_x}, Gradient_y = {grad_y}")

        if abs(new_x - x) < tolerance and abs(new_y - y) < tolerance:
            break

        x, y = new_x, new_y

    return x, y
initial_x, initial_y = 1.0, 1.0
learning_rate = 0.01
tolerance = 1e-6
max_iterations = 10000
optimal_x, optimal_y = gradient_descent(initial_x, initial_y, learning_rate, tolerance, max_iterations)
print("The minimum values of x and y are approximately:", optimal_x, optimal_y)
```

(1 ponto) Write Python code without using any libraries to find the value of x at which the
 sigmoid function z(x) shown in equation (3) has minimum value. Consider Gradient Descent
 Algorithm.
 z(x) = 1
 (3)
 1 +eâˆ’

```python
def sigmoid(x):
    return 1 / (1 + 2.71828 ** -x)

def derivative(x):
    sig_x = sigmoid(x)
    return sig_x * (1 - sig_x)
def gradient_descent(initial_x, learning_rate, tolerance, max_iterations):
    x = initial_x
    for iteration in range(max_iterations):
        grad = derivative(x)
        new_x = x - learning_rate * grad

        print(f"Iteration {iteration + 1}: x = {new_x}, Gradient = {grad}")

        if abs(new_x - x) < tolerance:
            break

        x = new_x

    return x

initial_x = 1.0
learning_rate = 0.1
tolerance = 1e-6
max_iterations = 10000
optimal_x = gradient_descent(initial_x, learning_rate, tolerance, max_iterations)
print("The minimum value of x is approximately:", optimal_x)
```

 (1 ponto) Write Python code without using any libraries to find the value of optimal values of
 model parameters M and C such that the modelâ€™s Square Error Value shown in equation 4 will
 be minimum. It means model gives output close to expected output as shown in Figure

```python
def compute_gradients(x, y, M, C):
    n = len(x)
    error_sum_M = sum((M * x[i] + C - y[i]) * x[i] for i in range(n))
    error_sum_C = sum((M * x[i] + C - y[i]) for i in range(n))
    dM = (2 / n) * error_sum_M
    dC = (2 / n) * error_sum_C
    return dM, dC

def gradient_descent(x, y, learning_rate, tolerance, max_iterations):
    M, C = 0.0, 0.0
    for iteration in range(max_iterations):
        dM, dC = compute_gradients(x, y, M, C)
        new_M = M - learning_rate * dM
        new_C = C - learning_rate * dC

        print(f"Iteration {iteration + 1}: M = {new_M}, C = {new_C}, Gradient_M = {dM}, Gradient_C = {dC}")

        if abs(new_M - M) < tolerance and abs(new_C - C) < tolerance:
            break

        M, C = new_M, new_C

    return M, C
x = [1, 2, 3, 4, 5]
y = [2, 4, 5, 4, 5]
learning_rate = 0.01
tolerance = 1e-6
max_iterations = 10000
optimal_M, optimal_C = gradient_descent(x, y, learning_rate, tolerance, max_iterations)
print("Optimal values: M =", optimal_M, ", C =", optimal_C)
```
